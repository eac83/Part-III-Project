```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import functions as fn
import my_rc_params
```

```python
df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'box_total_z0.csv')
k_tot_0 = np.array(df.iloc[:,0])
p_tot_0 = np.array(df.iloc[:,1])

df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'box_total_z2.csv')
k_tot_2 = np.array(df.iloc[:,0])
p_tot_2 = np.array(df.iloc[:,1])

df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'lin_z0.csv')
k_lin_0 = np.array(df.iloc[:,0])
p_lin_0 = np.array(df.iloc[:,1])

df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'lin_z2.csv')
k_lin_2 = np.array(df.iloc[:,0])
p_lin_2 = np.array(df.iloc[:,1])

df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'nonlin_z0.csv')
k_non_0 = np.array(df.iloc[:,0])
p_non_0 = np.array(df.iloc[:,1])

df = pd.read_csv(fn.HERE_PATH/'Power_Spectra'/'nonlin_z2.csv')
k_non_2 = np.array(df.iloc[:,0])
p_non_2 = np.array(df.iloc[:,1])
```

```python
fg, ax = plt.subplots()
ax.loglog()

f = lambda x: 2*np.pi/x
ax.tick_params(which='both', top=False)
Ax = ax.secondary_xaxis('top', functions=(f,f))
Ax.set_xlabel(r'$r\:[h^{-1}\mathrm{Mpc}]$')

ax.set_xlabel(r'$k\:[h\mathrm{Mpc}^{-1}]$')
ax.set_ylabel(r'$P_\mathrm{total}(k)[(h^{-1}\mathrm{Mpc})^{3}]$')
ax.plot(k_tot_2, p_tot_2, lw=3, c='b', label=2)
ax.plot(k_tot_0, p_tot_0, lw=3, c='r', label=0)
ax.legend(title=r'$z$')
fg.savefig(fn.HERE_PATH/'Power_Spectrum_Plots'/'Pylians.pdf', bbox_inches='tight')
```

```python
fg2, ax2 = plt.subplots()
lines1 = []
lines2 = []
ax2.loglog()
ax2.set_xlabel(r'$k\:[h\mathrm{Mpc}^{-1}]$')
ax2.set_ylabel(r'$P_\mathrm{total}(k)[(h^{-1}\mathrm{Mpc})^{3}]$')
lines1 += ax2.plot(k_tot_2, p_tot_2, lw=3, c='b', label=2)
lines2 += lines1[:]
lines1 += ax2.plot(k_tot_0, p_tot_0, lw=3, c='r', label=0)
lines2 += ax2.plot(k_lin_2, p_lin_2, '--', lw=3, c='b')
ax2.plot(k_lin_0, p_lin_0, '--', lw=3, c='r')
lines2 += ax2.plot(k_non_2, p_non_2, ':', lw=3, c='b', label='Total')
ax2.plot(k_non_0, p_non_0, ':', lw=3, c='r')
lg1 = ax2.legend(handles=lines1, title=r'$z$')
ax2.legend(handles=lines2, labels=['Total', 'Linear', 'Non-Linear'], loc='lower left')
ax2.add_artist(lg1)
f = lambda x: 2*np.pi/x
ax2.tick_params(which='both', top=False)
Ax2 = ax2.secondary_xaxis('top', functions=(f,f))
Ax2.set_xlabel(r'$r\:[h^{-1}\mathrm{Mpc}]$')

ax2.set_xlim(0.1, 100)
ax2.set_ylim(1e-4, 1e4)
fg2.savefig(fn.HERE_PATH/'Power_Spectrum_Plots'/'CAMB.pdf', bbox_inches='tight')
```
